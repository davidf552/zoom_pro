```python
import pandas as pd
import duckdb
```

```python
df = pd.read_csv("Video_Games_Sales_Cleaned.csv")
```

```python
sql_query = "SELECT console, ROUND(SUM(total_sales), 2) AS year_sales " \
            "FROM df GROUP BY console ORDER BY year_sales DESC"
```

```python
sql_query2 = "SELECT release_year, ROUND(SUM(total_sales), 2) AS year_sales " \
            "FROM df GROUP BY release_year ORDER BY release_year"
```

```python
duckdb.sql(sql_query2).write_csv("sales_by_year.csv")
```
